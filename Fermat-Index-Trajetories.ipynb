```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Update matplotlib settings to use LaTeX for text rendering
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"]
})

# Constants and range definitions
k = 3  # Constant used in the refractive index calculation
x = np.linspace(-5, 5, 100)  # x-range
y = np.linspace(0, 10, 100)  # y-range
X, Y = np.meshgrid(x, y)  # Create a meshgrid for x and y values

# Calculate the refractive index based on the provided formula
N = k / Y

# Set up the figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot a surface graph of the refractive index
surf = ax.plot_surface(Y, X, N, cmap='viridis')

# Set axis labels and the plot title
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_zlabel(r'$n(x,y)$')
ax.set_title(r'Refractive Index $n = \frac{k}{y}$, with $k=3$')

# Add a color bar to the plot to indicate scale of the refractive index
fig.colorbar(surf, shrink=0.5, aspect=5)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Update matplotlib settings to enable LaTeX formatting
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"]
})

# Generate a meshgrid for x and y coordinates
x = np.linspace(-5, 5, 100)  # x ranges from -5 to 5
y = np.linspace(0, 10, 100)  # y ranges from 0 to 10
X, Y = np.meshgrid(x, y)  # Create the meshgrid

# Calculate the refractive index n for each point on the meshgrid
N = np.sqrt(1 + Y)  # n is defined as the square root of (1 + y)

# Initialize a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface representing the refractive index
surf = ax.plot_surface(X, Y, N, cmap='viridis')  # Use the 'viridis' colormap

# Set the labels and title of the plot
ax.set_xlabel(r'$x$')  # x-axis label
ax.set_ylabel(r'$y$')  # y-axis label
ax.set_zlabel(r'$n(x,y)$')  # z-axis label, showing the refractive index
ax.set_title(r'Refractive Index $n = \sqrt{1 + y}$')  # Title of the plot

# Add a color bar to the plot to show the scale of the refractive index
fig.colorbar(surf, shrink=0.5, aspect=5)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Update matplotlib settings to enable LaTeX for text rendering
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"]
})

# Define the range of x values
x = np.linspace(-10, 10, 400)

# Constants and parameters for the circular trajectories
k = 3
A_values = [11, -7/2]  # Adjust the horizontal shift of the trajectories
B_values = [130/9, 125/36]  # Adjust the scale of the trajectories
colors = ['blue', 'red']  # Colors for plotting each trajectory
points = [[[-2, 7], [0, 3]], [[1, 5], [6, 5]]]  # Coordinates of specific points to be annotated

# Plot each circular trajectory
for i in range(len(A_values)):
    y = np.sqrt(k**2 * B_values[i] - (x + A_values[i])**2)
    plt.plot(x, y, label=f"$y({points[i][0][0]})={points[i][0][1]}, y({points[i][1][0]})={points[i][1][1]}$", color=colors[i])
    plt.scatter([point[0] for point in points[i]], [point[1] for point in points[i]], color=colors[i])

# Set axis labels and plot title
plt.xlabel(r'$x$')  # Label for the x-axis
plt.ylabel(r'$y$')  # Label for the y-axis
plt.title(r'Circular Trajectories $y = + \sqrt{k^2B - (x+A)^2}$, with $k=3$')  # Title of the plot
plt.ylim((0, 15))  # Set the limits for the y-axis
plt.legend()  # Display the legend for the plot

# Enhance the plot with grid lines and axis lines
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)  # Horizontal axis line
plt.axvline(0, color='black', linewidth=0.5)  # Vertical axis line

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Update matplotlib settings to enable LaTeX for text rendering
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"]
})

# Function to calculate the refractive index n(x, y)
def refractive_index(x, y):
    return np.sqrt(1 + y**2)

# Generate the range of x values
x = np.linspace(-2, 2, 200)
# Calculate exponential functions for y values
y1 = np.exp(-x)
y2 = -np.exp(-x)
y3 = np.exp(x)
y4 = -np.exp(x)

# Plotting the exponential curves
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label=r'$y = e^{-x}$')
plt.plot(x, y2, label=r'$y = -e^{-x}$')
plt.plot(x, y3, label=r'$y = e^{x}$')
plt.plot(x, y4, label=r'$y = -e^{x}$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.title('Light Trajectories on a Flat Plate')
plt.legend()
plt.grid(True)
plt.show()

# Generate a grid of points in the (x, y) plane
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)

# Calculate the refractive index at each point on the grid
n_vals = refractive_index(X, Y)

# Plotting the refractive index distribution in 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, n_vals, cmap='viridis')
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_zlabel(r'$n(x, y)$')
ax.set_title(r'Refractive Index $n = \sqrt{1 + y^2}$')

# Add a color bar to the plot to indicate the scale of the refractive index
fig.colorbar(surf, shrink=0.5, aspect=5)

# Display the plot
plt.show()
```
